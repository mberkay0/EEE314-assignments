```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.offline as py
import warnings
warnings.filterwarnings('ignore')
import plotly
import plotly.graph_objs as go
import plotly.express as ex
from plotly.subplots import make_subplots
import plotly.express as px
```

```python
!unzip /content/happiness_index.zip
```

```python
df21 = pd.read_csv("/content/world-happiness-report-2021.csv")
df_up_to20 = pd.read_csv("/content/world-happiness-report.csv")
```

```python
df21.rename(columns={"Country name": "Country",
                     "Healthy life expectancy":"Life Expectancy",
                     "Freedom to make life choices":"Freedom",
                     "Perceptions of corruption":"Corruption",
                     "Logged GDP per capita": "GDP per capita"},inplace=True)

df21.drop(["Explained by: Log GDP per capita",
           "Explained by: Social support",
           "Explained by: Healthy life expectancy",
           "Ladder score in Dystopia",
           "Explained by: Freedom to make life choices",
           "Explained by: Generosity",
           "Explained by: Perceptions of corruption",
           "Dystopia + residual",
           "Standard error of ladder score",
           "upperwhisker","lowerwhisker"], axis=1,inplace=True)
df21['year'] = 2021
df21
```

```python
df_up_to20.rename(columns={"Country name": "Country",
                           "Healthy life expectancy at birth": "Life Expectancy",
                           "Freedom to make life choices": "Freedom",
                           "Perceptions of corruption": "Corruption",
                           "Log GDP per capita": "GDP per capita",
                           "Life Ladder":"Ladder score"}, inplace=True)
df_up_to20.drop(["Positive affect", "Negative affect"], axis=1, inplace=True)
country_region_data = df21[["Country", "Regional indicator"]]
df_up_to20 = df_up_to20.merge(country_region_data, how='left', left_on='Country', right_on='Country')
df_up_to20
```

```python
data = pd.concat([df_up_to20, df21])
data = data.astype({'year': 'str'})
data
```

```python
region_list = []
null = data[data["Regional indicator"].isnull()]

for country in data['Country']:
    if country == "Angola":
        region_list.append("Sub-Saharan Africa")
    if country == "Belize":
        region_list.append("Latin America and Caribbean")
    if country == "Bhutan":
        region_list.append("South Asia")
    if country == "Central African Republic":
        region_list.append("Sub-Saharan Africa")
    if country == "Congo (Kinshasa)":
        region_list.append("Sub-Saharan Africa")
    if country == "Cuba":
        region_list.append("Latin America and Caribbean")
    if country == "Djibouti":
        region_list.append("Middle East and North Africa")
    if country == "Guyana":
        region_list.append("Latin America and Caribbean") 
    if country == "Oman":
        region_list.append("Middle East and North Africa")  
    if country == "Qatar":
        region_list.append("Middle East and North Africa") 
    if country == "Somalia":
        region_list.append("Middle East and North Africa") 
    if country == "Somaliland region":
        region_list.append("Sub-Saharan Africa") 
    if country == "South Sudan":
        region_list.append("Sub-Saharan Africa") 
    if country == "Sudan":
        region_list.append("Middle East and North Africa")  
    if country == "Suriname":
        region_list.append("Latin America and Caribbean") 
    if country == "Syria":
        region_list.append("Middle East and North Africa") 
    if country == "Trinidad and Tobago":
        region_list.append("Latin America and Caribbean")     
        
        
for i,j in zip(null.index,range(len(null))):
    data["Regional indicator"][i] = region_list[j]
```

```python
data.isnull().sum()
```

```python
for column in data.columns:
    if data[column].dtype != "float64" :
        pass
    else:
        data[column] = data[column].fillna(data[column].mean())  
```

```python
data = data.dropna()
data.isnull().sum()
```

```python
q25 = df21['Ladder score'].quantile(.25)
q75 = df21['Ladder score'].quantile(.75)
mean = df21['Ladder score'].mean()

unhappiest = df21[df21['Ladder score'] < q25]
normal = df21[(df21['Ladder score'] > q25) & (df21['Ladder score'] < q75)]
happiest = df21[df21['Ladder score'] > q75]

unhappiest=unhappiest.groupby('Regional indicator').count().reset_index()
normal=normal.groupby('Regional indicator').count().reset_index()
happiest=happiest.groupby('Regional indicator').count().reset_index()

fig = make_subplots(rows=1, cols=3, shared_yaxes=True)
traces_names = ['unhappy', 'normal', 'happy']
colors = ['rgba(255, 65, 54, 0.5)', 'rgba(255, 144, 14, 0.5)', 'rgba(44, 160, 101, 0.5)']
categorise_data = [unhappiest, normal, happiest]

for index, categorise in enumerate(categorise_data):
    y = [i for i in categorise['Country']]
    x = [i for i in categorise['Regional indicator']]
    fig.add_trace(go.Bar(x=x, 
                         y=y, 
                         name=traces_names[index],
                         marker_color=colors[index],
                         text=y), 1, index+1)
    
fig.update_traces(texttemplate='%{text:}', textposition='outside')

fig.update_layout(
    title='Number of countries regionally Happiness Categorisation',
    title_font_size=18,
    yaxis_title="Count of Country")

fig['layout']['xaxis2']['title']='Regions'

fig.show()
```

Dünya genelinde kategorize edilen 3 grup arasında ortalama öznitelik değerlerine bakalım ve ilişkilendirmeye çalışalım.

Bu analizde bölgeler arasında freedom, generosity ve corruption değerlerinin ortalama olarak aynı değerlerde olduğu görülmektedir. Fakat mutlu ülkeler ve bölgelerin yaşam standartlarının daha yüksek olduğu daha uzun yaşam ömrüne sahip oldukları açıktır.

```python
fig = go.Figure()

for index, categorise in enumerate(categorise_data):
    x = [i for i in categorise.columns if i != 'Country' and i != 'Regional indicator']
    y = [i for i in categorise.mean()]
    fig.add_trace(go.Scatter(x=x, 
                         y=y, 
                         name=traces_names[index],
                         marker_color=colors[index],
                         text=y))

fig.update_layout(
    title='Average attribute values among 3 groups categorized worldwide',
    title_font_size=18,
    yaxis_title="Average Scores",
    xaxis_title="Features")

fig.show()
```
