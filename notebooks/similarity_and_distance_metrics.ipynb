Öklid Mesafesi formülü:

$d\left( x,y\right)   = \sqrt {\sum _{i=1}^{n}  \left( x_{i}-y_{i}\right)^2 } $

Öklid mesafesi, birçok makine öğrenimi algoritması için temel oluşturan en yaygın kullanılan metriklerden biridir. Veriler yoğun veya sürekli olduğunda, bu metrik en iyi yakınlık ölçüsüdür. Bununla birlikte, çok sayıda özellik içeren çok büyük veri kümeleriyle karşılaşıldığında, bu metrik hafıza hatalarına yol açabilir. İki nokta arasındaki Öklid mesafesi, onları birbirine bağlayan yolun uzunluğudur. İki nokta arasındaki bu mesafe Pisagor teoremi ile verilmektedir. L2 norm mesafe metriği olarak da bilinir. En popüler mesafe metriklerindendir.

```python
def euclidean_distance(u, v):
    """This function calculates 
    the Euclidean distance between 
    two given vectors (u, v).

    Args:
        u ([list]): Input array.
        v ([list]): Input array.

    Returns:
        distance: Calculated Euclidean distance.
    """
    distance = sqrt(sum((a-b)**2 for a, b in zip(u, v)))
    return distance
```

Kosinüs benzerliği formülü:

$\cos(\varphi) = Cos.Sim. ({\bf A},{\bf B})= {{\bf A} . {\bf B} \over \|{\bf A}\| \|{\bf B}\|} = \frac{ \sum_{i=1}^{n}{{\bf A}_i{\bf B}_i} }{ \sqrt{\sum_{i=1}^{n}{({\bf A}_i)^2}} \sqrt{\sum_{i=1}^{n}{({\bf B}_i)^2}} }$


Kosinüs benzerlik metriği, iki özelliğin normalleştirilmiş iç çarpımını bulur. 0 ° 'nin kosinüsü 1'dir ve diğer herhangi bir açı için 1'den küçük olarak bulunur. Dolayısıyla, büyüklük değil yönelim yargısıdır: aynı yönelimdeki iki vektörün kosinüs benzerliği 1, 90 ° 'deki iki vektörün benzerliği 0 ve taban tabana zıt iki vektörün benzerliği ise -1'dir. Kosinüs benzerliği özellikle sonucun düzgün bir şekilde [0,1] ile sınırlandığı pozitif boşlukta kullanılır. Kosinüs benzerliğinin popülaritesinin nedenlerinden biri, özellikle seyrek vektörler için değerlendirmenin çok verimli olmasıdır.
Kosinüs metriği, kullanıcılara gelecekteki önerileri sunmak için temel olarak İşbirlikçi Filtreleme tabanlı öneri sistemlerinde kullanılır. Ayrıca doğal dil işleme ve metin benzerliği konularında sıklıkla kullanılmaktadır. Çok sık olmamakla birlikte görüntü işleme problemlerinin çözümlerinde de kullanılmaktadır.

<div>


</div>

```python
def cosine_similarity(u, v):
    """
    Compute cosine similarity between samples in u and v.

    Args:
        u ([list]): Input array.
        v ([list]): Input array.

    Returns:
        similarity: Calculated cosine similarity.
    """
    dot_product = sum(a*b for a, b in zip(u, v))
    norm_a = sum(a*a for a in u) ** 0.5
    norm_b = sum(b*b for b in v) ** 0.5
    similarity = dot_product / (norm_a * norm_b) 
    
    return similarity
```

```python
cosine_similarity(u, v)
```

Korelasyon formülü:

${\displaystyle \rho _{X,Y}=\operatorname {corr} (X,Y)={\operatorname {cov} (X,Y) \over \sigma _{X}\sigma _{Y}}={\operatorname {E} [(X-\mu _{X})(Y-\mu _{Y})] \over \sigma _{X}\sigma _{Y}}}$

İki rassal değişken arasındaki doğrusal ilişkinin yönünü ve gücünü belirtir. Genel istatistiksel kullanımda korelasyon, bağımsızlık durumundan ne kadar uzaklaşıldığını gösterir.Korelasyon katsayısı, bağımlı değişken ile bağımsız değişkenler arasındaki ilişkinin gücünü gösteren bir katsayıdır. Örneğin; öğrencinin ders çalışma süresi ile aldığı not ile arasında bir ilişki olup olmadığı korelasyon katsayısı ile incelenebilir. Korelasyon katsayısı değişkenlerin yönü ve etkileşimlerin nasıl olduğu hakkında bilgi verir.
Korelasyon -1 ≤ corr ≤1 aralığındadır.

Korelasyon hesabı için gerekli matematiksel ifadeler:

$ mean(k) = μ_{k}={1\over n} \sum_{i=1}^{n} k_{i} $


${stdev}(k)  = \sqrt{{1\over n-1}\sum_{i=1}^n (k_i - \bar{k})^2}$


${\displaystyle \operatorname {cov} (X,Y)={\frac {1}{n}}\sum _{i=1}^{n}(x_{i}-μ_X)(y_{i}-μ_Y).}$

![image.png](https://upload.wikimedia.org/wikipedia/commons/0/02/Correlation_examples.png)

Korelasyon katsayısının 0’a yaklaşması değişkenler arasında zayıf bir ilişkinin olduğunu gösterir. Değişkenlerin her ikiside artıyor veya azalıyorsa pozitif yönde, değişkenlerden biri artarken diğeri azalıyor ise negatif yönde bir ilişki gösterir.

```python
def mean(k):
    """
    Compute the arithmetic mean.

    Args:
        k ([list]): Input array.

    Returns:
        mean: Calculated mean.
    """
    n = len(k)
    mean = sum(k) / n
    return mean 


def std_dev(k):
    """
    Compute the standard deviation.

    Args:
        k ([list]): Input array.

    Returns:
        std: Calculated standard deviation.
    """
    mean_k = mean(k)
    var  = sum(pow(i-mean_k, 2) for i in k) / len(k)  
    std  = sqrt(var)  
    return std


def covariance(x, y):
    """
    Compute the covariance between samples in x and y.

    Args:
        x ([list]): Input array.
        y ([list]): Input array.

    Returns:
        cov: Calculated covariance.
    """
    mean_x = mean(x)
    mean_y = mean(y)
    cov = sum(((xk-mean_x)*(yk-mean_y)/len(x)) for xk, yk in zip(x, y))
    return cov


def correlation(x, y):
    """
    Compute the correlation between samples in x and y.

    Args:
        x ([list]): Input array.
        y ([list]): Input array.

    Returns:
        corr: Calculated correlation coefficient.
    """
    corr = covariance(x, y) / (std_dev(x)*std_dev(y))
    return corr
```

```python
correlation(x, y)
```

Manhattan Mesafesi formülü:

${\displaystyle d_{1}(\mathbf {x} ,\mathbf {y} )=\|\mathbf {x} -\mathbf {y} \|_{1}=\sum _{i=1}^{n}|x_{i}-y_{i}|}$


Manhattan mesafesi, iki nokta arasındaki mesafenin, Kartezyen koordinatlarının mutlak farklarının toplamı olduğu bir metriktir. Basit bir deyişle, x koordinatları ve y koordinatları arasındaki farkın mutlak toplamıdır. Bir Nokta A ve bir Nokta B'ye sahip olduğumuzu varsayalım: eğer aralarındaki Manhattan mesafesini bulmak istiyorsak, sadece mutlak x ekseni ve y ekseni değişimini toplamamız gerekir. İki nokta arasındaki Manhattan mesafesini, eksenler boyunca dik açılarla ölçerek buluruz. Bu metrik: doğrusal mesafe, L1 normu, şehir bloğu mesafesi ve Minkowski L1 mesafesi olarak da bilinir.

<div>

</div>

Izgara benzeri bir yoldaki iki veri noktası arasındaki mesafe hesaplanması gerekirse genellikle bu metrik kullanılır. D boyutluluğunun sabit (yüksek) bir değerine sahip belirli bir problem için, daha düşük p değerleri kullanmak tercih edilebilir. Bu, yüksek boyutlu uygulamalar için en çok tercih edilen L1 mesafe metriğinin (Manhattan Mesafe metriği) olduğu anlamına geliyor. "[On the Surprising Behavior of Distance Metrics
in High Dimensional Space](https://bib.dbvis.de/uploadedFiles/155.pdf)" Bu nedenle veride yüksek boyutluluk olduğunda Manhattan mesafesi genellikle daha yaygın olan Öklid mesafesine tercih edilir.
Tipik kullanım durumları: Veriler bilgisayar ağları üzerinden iletildiğinde hata düzeltme / algılama içerir. Hatayı tahmin etmenin bir yolu olarak ikili bir sözcükteki bozulmuş bitlerin sayısını belirlemek için kullanılabilir.
Ayrıca, kategorik değişkenler arasındaki mesafeyi ölçmek için Hamming mesafesini de kullanabilirsiniz. Büyüklük önemli bir ölçü olduğunda bu mesafe ölçüsünün kullanılması tavsiye edilmez. Veri kümesi ayrık ve / veya ikili özniteliklere sahip olduğunda Öklit mesafesine göre daha iyi sonuç verebilir.

```python
def manhattan_distance(u, v):
    """This function calculates 
    the Manhattan distance between 
    two given vectors (u, v).

    Args:
        u ([list]): Input array.
        v ([list]): Input array.

    Returns:
        distance: Calculated Manhattan distance.
    """
    distance = sum(abs(a-b) for a, b in zip(u, v))
    return distance
```

```python
manhattan_distance(u, v)
```

Minkowski Mesafesi formülü:

${\displaystyle D\left(X,Y\right)=\left(\sum _{i=1}^{n}|x_{i}-y_{i}|^{p}\right)^{1/p}}$

Minkowski mesafesi, Öklid mesafesi ve Manhattan mesafesinin genelleştirilmiş bir metrik biçimidir. Yani;

p = 2, Minkowsky mesafesi Öklid mesafesi ile aynıdır.

p = 1, Minkowsky mesafesi Manhattan mesafesi ile aynıdır.


Ara değerler, iki ölçü arasında kontrollü bir denge ve mesafe ölçümü üzerinde büyük miktarda esneklik sağlar.

Ayarlanabilen bir hiperparametre " p " aracılığıyla gerçek değerli vektörler için kullanılan mesafe ölçümü türü üzerinde kontrol sağladığından, mesafe ölçümlerini kullanan bir makine öğrenimi algoritması uygularken Minkowski mesafesinin kullanılması yaygındır .


<div>

</div>

"Normlu vektör uzayı, üzerinde bir normun tanımlandığı bir vektör uzayıdır." [Vikipedia'ya göre](https://en.wikipedia.org/wiki/Minkowski_distance)

Minkowski mesafesi, değişkenler mutlak sıfır değeriyle oran ölçeklerinde ölçüldüğünde sıklıkla kullanılır. Daha geniş bir aralığa sahip değişkenler, sonucun önüne geçebilir. Yüksek değerlere sahip birkaç aykırı değer bile sonucu saptırır ve daha düşük bir üst sınıra sahip birkaç değişken tarafından verilen benzerliği göz ardı eder.

```python
def minkowski_distance(u, v, p):
    """This function calculates 
    the Minkowski distance between 
    two given vectors (u, v).

    Args:
        u ([list]): Input array.
        v ([list]): Input array.
        p ([float]): Minkowski p_value (Ex: 1, 1.1, 1.8 ...)

    Returns:
        distance: Calculated Minkowski distance.
    """
    distance = sum(pow(abs(a-b), p) for a, b in zip(u, v))** (1/p)
    return distance
```

Eğer p = 1 olarak ayarlanırsa bu metrik bize Minkowski mesafesini verir.

```python
minkowski_distance(u, v, p=1) #  the Minkowski distance with p set to 1 to give the Manhattan distance
```

Eğer p = 2 olarak ayarlanırsa bu metrik bize Öklit mesafesini verir.

```python
minkowski_distance(u, v, p=2) # p set to 2 to give the Euclidean distance
```

Hamming mesafesi, iki ikili veri dizisini karşılaştırmak için bir ölçüdür. Eşit uzunluktaki iki ikili dizgiyi karşılaştırırken, Hamming mesafesi, iki bitin farklı olduğu bit konumlarının sayısıdır.
İki dizge arasındaki Hamming mesafesi, a ve b, d (a, b) olarak belirtilir.
İki dizge arasındaki Hamming mesafesini hesaplamak için ve XOR işlemini gerçekleştiririz (a⊕ b) ve sonra sonuç dizisindeki toplam 1 sayısını sayarız.
11011001 ve 10011101 olmak üzere iki dizi olduğunu varsayalım.
11011001 ⊕ 10011101 = 01000100. Bu iki 1'i içerdiğinden, Hamming mesafesi, d (11011001, 10011101) = 2.

Hamming mesafesi, bilgi teorisi , kodlama teorisi ve kriptografi dahil olmak üzere birçok disiplinde kullanılmaktadır. Telekomünikasyonda, sabit uzunluklu ikili (binary) bir sözcükteki çevrilmiş bitlerin sayısını bir hata tahmini olarak saymak için kullanılır ve bu nedenle bazen sinyal mesafesi olarak da adlandırılır.

```python
def hamming_distance(a, b):
    """Compute the Hamming distance 
    between equal-length sequences.

    Args:
        u ([list]): Input array.
        v ([list]): Input array.

    Returns:
        distance: Calculated Hamming distance.
    """
    distance = sum(s1 != s2 for s1, s2 in zip(a, b))
    return distance
```

Jaccard benzerliği formülü:


<div>
<img src="https://upload.wikimedia.org/wikipedia/commons/c/c7/Intersection_over_Union_-_visual_equation.png" width="200"/>
</div>


${\displaystyle J(A,B)={{|A\cap B|} \over {|A\cup B|}}={{|A\cap B|} \over {|A|+|B|-|A\cap B|}}}$

Jaccard benzerliği, sonlu örnek kümeleri arasındaki benzerliği ölçer ve kümelerin kesişme noktasının, örnek kümelerinin birleşiminin esas niteliğine bölünmesi olarak tanımlanır.

A ve B kümeleri arasında Jaccard benzerliği bulmak istediğinizi varsayalım, bu A ∩ B ve A ∪ B'nin önemliliğinin oranıdır. Yani basit bir ifadeyle elimizde bulunan iki kümenin kesişiminin elaman sayısının, birleşiminin elaman sayısına bölümü şeklinde ifade edilebilir.

Bu benzerlik ölçütü makine öğrenimi ve derin öğrenmenin neredeyse her alanında kullanılır. Doğal dil işleme, görüntü işleme, anlamsal bölütleme gibi çoğu alanda kullanılmaktadır. Jaccard indeksi genellikle ikili verilerin kullanıldığı uygulamalarda kullanılır. Bir görüntünün segmentlerini, örneğin bir arabayı tahmin eden bir derin öğrenme modeliniz olduğunda, Jaccard indeksi, gerçek etiketler verildiğinde tahmin edilen segmentin ne kadar doğru olduğunu hesaplamak için kullanılabilir. Benzer şekilde, metin benzerlik analizinde kullanılabilir.

<div>
<img src="https://divingintogeneticsandgenomics.rbind.io/img/headers/scclusteval_flow.png" width="700"/>
</div>


```python
def jaccard_similarity(u, v):
    """Compute the Jaccard similarity 
    between 2 sequences.

    Args:
        u ([list]): Input array.
        v ([list]): Input array.

    Returns:
        similarity: Calculated Jaccard similarity.
    """
    intersection = len(list(set(u).intersection(v)))
    union = (len(u) + len(v)) - intersection
    similarity = float(intersection) / union
    return similarity
```

##Dice Benzerliği (Dice Similarity)


Öncelikle bir özellik çıkarımı yöntemi ile iki veri üzerinden özellikler çıkarılır ve ardından aşağıdaki formüle göre benzerlik hesabı yapılır.

Dice benzerliği formülü:

${\displaystyle DSC={\frac {2|X\cap Y|}{|X|+|Y|}}}$

Sørensen endeksi de denilen bu metrik, her iki küme için ortak olan öğe sayısının iki katına eşittir ve her kümedeki öğe sayısının toplamına bölünür. Boolean verilerine uygulandığında, gerçek pozitif (TP), yanlış pozitif (FP) ve yanlış negatif (FN) tanımları kullanılarak şu şekilde yazılabilir: 

${\displaystyle DSC={\frac {2TP}{2TP+FP+FN}}}$

Hem payda hem de paydada yalnızca bir kez gerçek pozitifleri sayan Jaccard endeksinden farklıdır . DSC, 0 ile 1 arasında değişir. Kümeler üzerindeki benzerlik ölçüsü olarak görülebilir.

Bu benzerlik ölçütü mesafe metriği olarak ele alınmak istenirse:

${\displaystyle d=1-{\frac {2|X\cap Y|}{|X|+|Y|}}}$

şeklinde bulunabilir. Öklid mesafesi ile karşılaştırıldığında , Sørensen mesafesi daha heterojen veri setlerinde hassasiyeti korur ve aykırı değerlere daha az ağırlık verir.

Bu metrik doğal dil işlemeden tutun makina öğrenimi ve derin öğrenme alanlarının birçoğunda kullanılmaktadır. Çok popüler bir metriktir ve varyasyonları çoktur. Ayrıca genom ve metagenom mesafe tahmini için Mash mesafesinin bir parçası olarak kullanılmaktadır.

```python
def dice_coef(u, v):
    """Compute the Jaccard similarity 
    between 2 sequences.

    Args:
        u ([list]): Input array.
        v ([list]): Input array.

    Returns:
        dsc: Calculated Dice similarity coefficient.
    """
    intersection = len(list(set(u).intersection(v)))
    dsc = (2*intersection) / (len(u) + len(v))
    return dsc 
```

Chebyshev mesafesi formülasyonları:

${\displaystyle D_{\rm {Chebyshev}}(x,y):=\max_{i}(|x_{i}-y_{i}|)\ }$

${\displaystyle \lim _{p\to \infty }{\bigg (}\sum _{i=1}^{n}\left|x_{i}-y_{i}\right|^{p}{\bigg )}^{1/p}}$

${\displaystyle D_{\rm {Chebyshev}}=\max \left(\left|x_{2}-x_{1}\right|,\left|y_{2}-y_{1}\right|\right)}$

Satrançda şah, bir adımda 8 bitişik kareden herhangi birine gidebilir (Aşağıdaki şekil üzerinden görebilirsiniz.).

<div>

</div>

Öyleyse kralın (x1, y1) kareden (x2, y2) kareye kaç adım gitmesi gerekir? Eğer denerseniz minimum adım sayısının her zaman maksimum (| x2-x1 |, | y2-y1 |) adım olduğu görülecektir. Chebyshev mesafesi (L∞ normu) olarak adlandırılan benzer bir mesafe ölçüm yöntemi vardır.

Chebyshev mesafe metriği, Bilgisayar Destekli Üretim (Computer-aided manufacturing) uygulamalarında, özellikle bunlara yönelik optimizasyon algoritmalarında yaygın olarak kullanılmaktadır. (Düzlemde çalışan plotlama veya delme makineleri, fotoplotter vb. gibi birçok araç)
Ayrıca Chebyshev mesafesi, bir vincin bir nesneyi hareket ettirmek için harcadığı süreye çok benzediğinden, depo lojistiğinde sıklıkla kullanılır.
 



```python
def chebyshev_distance(u, v):
    """Compute the Chebyshev distance 
    between 2 sequences.

    Args:
        u ([list]): Input array.
        v ([list]): Input array.

    Returns:
        distance: Calculated Chebyshev distance.
    """
    distance = max(abs(a - b) for a, b in zip(u, v))
    return distance
```
